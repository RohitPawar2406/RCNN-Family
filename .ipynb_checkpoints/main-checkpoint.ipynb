# Implementing SelectiveSearch to generate region proposals

```python
img.shape,1200*1600
```

```python
def extract_candidate(img):
    
#     im_orig : ndarray
#         Input image
#     scale : int
#         Free parameter. Higher means larger clusters in felzenszwalb segmentation.
#     sigma : float
#         Width of Gaussian kernel for felzenszwalb segmentation.
#     min_size : int
#         Minimum component size for felzenszwalb segmentation.
    img_lbl, regions = selectivesearch.selective_search(img,scale=200,min_size=100)
    img_area = np.prod(img.shape[:2])
    
    # regions
#     [{'rect': (0, 0, 17, 63), 'size': 830, 'labels': [0.0]},
#      {'rect': (16, 0, 342, 462), 'size': 58610, 'labels': [1.0]},
#      {'rect': (50, 0, 60, 69), 'size': 246, 'labels': [2.0]}]
    
    
    #np.prod(img.shape[:2]) #Image Area
    candidates = []
    
    for r in regions:
        if r['rect'] in candidates:
            continue
        if r['size'] < (0.05*img_area):
            continue
        if r['size'] > (1*img_area):
            continue
        
        x,y,w,h = r['rect']
        candidates.append(list(r['rect']))
    
    return candidates
```

```python
a= extract_candidate(img)
```

```python
len(a) # 180 regions proposals
```

```python
show(img, bbs=a)
```

# Intersection within the term Intersection over Union measures how overlapping the predicted and actual bounding boxes are, while Union measures the overall space possible for overlap. IoU is the ratio of the overlapping region between the two bounding boxes over the combined region of both the bounding boxes.

```python
# For IoU use torchvision.ops
from torchvision.ops import box_iou
```

```python
a = torch.tensor([[0.0, 33.0,0.0, 50.0]])
b = torch.tensor([[50.0, 0.0, 60.0, 69.0]])
```

```python
return_Box = box_iou(a,b)
return_Box
```
